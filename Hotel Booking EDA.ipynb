## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Convert 'reservation_status_date' to datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
````

````python
# remove duplicates data
df.drop_duplicates(inplace = True)
df.shape
````

````python
# fill agent columns null values by 0
df['agent']=df['agent'].fillna(0)

# fill company columns null values by 0
df['company']=df['company'].fillna(0)

# fill country columns null values by undefined
df['country']=df['country'].fillna('UNDEFINED')

df.info()
````

````python
# drop remaining null values
df.dropna(inplace=True)
df.info()
````

````python
# Create a new column 'total_no_of_people' by summing the values of 'adults', 'children', and 'babies' columns
df['total_no_of_people'] = df['adults'] + df['children'] + df['babies']
````

````python
# Create a new column 'total_no_of_room_sold' by summing the values of 'stays_in_weekend_nights' and 'stays_in_week_nights' columns
df['total_no_of_room_sold'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
````

````python
# Create a new column 'revenue' by multiplying the 'total_no_of_room_sold' by the 'adr' (average daily rate) column
df['revenue'] = df['total_no_of_room_sold'] * df['adr']
````

````python
# Retrieve the unique values from the 'arrival_date_year' column
df['arrival_date_year'].unique()
````

````python
# Retrieve the unique values from the 'country' column
df['country'].unique()
````

````python
# find mean values
mean = df.adr.mean()

# find std values
std = df.adr.std()

# find postive outliers
pos_out = mean + 3*std

# find negative outliers
neg_out = mean - 3*std

print(pos_out, neg_out)

````

````python
# Filtering out rows where 'adr' is between 'neg_out' and 'pos_out'
# '~' is the negation operator, so it selects rows where the condition inside the brackets is False
dect_out = df[~((df['adr']>=neg_out) & (df['adr']<=pos_out))]

dect_out.shape
````

### What all manipulations have you done and insights you found?

* Modified the datatype of the reservation_status_date column object to Datetime
* Handle 31994 duplicate values
* Filled in the agent company and columns null values with 0 and country with UNDEFINED, respectively
* Eliminated the remaining null values
* Create a new column (total_no_of_people) by adding the sums of the columns for adults, children, and babies
* Create a new column (total_no_of_room_sold) by adding the sums of the stays_in_weekend_nights and stays_in_weeknights
* Create a new column (revenue) and multiply it by the sum of the no_of_room_sold and adr
* Discovered 628 outliers but left them all in place since there was no compelling reason to remove any of them.
